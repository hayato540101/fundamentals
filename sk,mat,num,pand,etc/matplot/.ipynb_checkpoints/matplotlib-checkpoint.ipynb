```python
import matplotlib.pyplot as plt
np.random.seed(123)
fig1, fig2 = plt.figure(1),plt.figure(2)
plt.figure(1)
plt.plot(np.random.rand(100))
plt.figure(2)
plt.scatter(np.random.rand(100),np.random.rand(100))
plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(5,5))
# FigureオブジェクトとAxesオブジェクトが表示されていることが分かる
# ax.shape #(2, 2)

print(fig.axes) # axesオブジェクトのリスト
print(fig.axes[1]) # 一応個別にアクセスできる

nrow,ncol = ax.shape
size = 100
for i in range(nrow):
    for j in range(ncol):
        x = np.random.rand(size)
        y = 2 * x + np.random.randn(size)
        ax[i,j].scatter(x,y)
        ax[i,j].set_title('[{},{}]'.format(i,j))
```

ここからわかるのは以下の関係です。AxesSubplotはplt.subplotsで作った場合のAxesオブジェクトと思って問題ありません。

Figureは自分の下のAxesを知っているけどその下のAxisは知らない
Axesは自分の上のFigureと下のAxisを知っている
Axisは自分の上のAxesとその上のFigureを知っている
Figureは複数のAxesを持てる（printで表示されているのが[]で囲まれたリストなので）
Axesは一つのFigureにしか属せない（printで表示されているのがリストではないので）
AxesはXAxisとYAxisを一つずつしか持てない（同上）
XAxisとYAxisは一つのAxes、一つのFigureにしか属せない

<img src = "components of a Matplotlib figure.png">

Artistはcontainer（容器）とprimitive（原始的なものという意味）の二種類に分類される。先の階層構造で示したFigure、Axes、Axisはcontainerに、プロットの線(Line2D)や点(PathCollection)あるいは文字(Text)はprimitiveに相当

<img src = "container,primitive.png">

グラフに図形やテキストなどのArtistを追加するときに、Figureに追加するかAxesに追加するかによって座標の原点が異なるからです。どちらを使うのが適切かはやりたいことや実装の方法によって変わる

```python
# 左右の軸に異なるスケールのプロットをした時に
# 素直にax.legendを使うと凡例が二つできていしまう

x = np.linspace(0, 2*np.pi, 100)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x), label='sin(x)')
ax1 = ax.twinx() # 生成したaxを使ってx軸が共通で異なるy軸のグラフを書く。
ax1.plot(x, 2*np.cos(x), c='C1', label='2cos(x)') 
'''legendの枠が2つできてしまう例'''
ax.legend()
ax1.legend()
```

```python
'''fig.legendでfigに属するaxesのlegendを一括表示できる
・デフォルトではFigureの座標が使われるが、凡例の配置はAxes座標ベース
の方が都合の良い場合が多い。下の例のようにbbox_transformオプションを
使うとaxに対する座標系に変更できる。
・バウンディングボックス（Bounding Box）とは，図形をちょうど囲うのに必要な大きさの，四角い箱 （矩形）のこと

'''

fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)
fig
```

```python
import matplotlib.ticker as ticker # FormatterとLocatorはTickerモジュールが必要

ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5*np.pi)) # 0.5*piごとにtickを書く
fig # jupyter notebookでは一度作ったFigureを再度表示させるにはfigを呼ぶ
```

```python
x = np.linspace(0, 2*np.pi, 100)

fig = plt.figure() # Figureを作成
ax = fig.add_subplot(1,1,1) # Axesを作成
print('ax.lines before plot:\n', ax.lines) # Axes.linesは空リスト
line1, = ax.plot(x, np.sin(x), label='1st plot') # Axes.linesにLine2Dを追加
print('ax.lines after 1st plot:\n', ax.lines)
line2, = ax.plot(x, np.sin(x+np.pi/8), label='2nd plot') # Axes.linesにLine2Dを追加
print('ax.lines after 2nd plot:\n', ax.lines)
print('ax.collections before scatter:\n', ax.collections)
scat = ax.scatter(x, np.random.rand(len(x)), label='scatter') # Axes.collectionsにPathCollectionを追加
print('ax.collections after scatter:\n', ax.collections)
ax.legend()
print('line1:', line1) # line2Dオブジェクトが追加されている
print('line2:', line2) # line2Dオブジェクトが追加されている
print('scat:', scat)
ax.set_xlabel('x value')
ax.set_ylabel('y value')
```
